```python
#Problem-4
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
A_inv = np.linalg.inv(A)
identity = np.dot(A,A_inv)
print("A * A^-1):-\n",identity)
AB = np.dot(A,B)
BA = np.dot(B,A)
print(AB)
print(BA)
```

```python
AB_T = np.dot(A,B).T
BT_AT = np.dot(B.T,A.T)
print(AB_T)
print(BT_AT)
```

```python
A = np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B = np.array([-1,-3,9])
X = np.dot(np.linalg.inv(A),B)
print(X)
X_solve = np.linalg.solve(A,B)
print(X_solve)
```
