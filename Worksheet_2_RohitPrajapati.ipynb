```python
#Problem_2
import pandas as pd
medical_data = pd.read_csv("C:/Users/Natsu Rohit/Downloads/medical_students_dataset.csv")
print(medical_data)
```

```python
print(medical_data.info())
```

```python
print("\n Columns with missing values:-")
print(medical_data.isnull().sum())
```

```python
numeric_cols = ["Age","Height","Weight","BMI","Temperature","Heart Rate","Cholesterol"]

for col in numeric_cols:
    medical_data[col] = medical_data[col].fillna(medical_data[col].mean())

cat_cols = ["Student ID","Gender","Blood Type","Blood Pressure","Diabetes","Smoking"]

for col in cat_cols:
    medical_data[col] = medical_data[col].fillna(medical_data[col].mode()[0])
```

```python
print(medical_data.isnull().sum())
```

```python
print("The number of duplicated rows is:-", medical_data.duplicated().sum())

medical_data = medical_data.drop_duplicates()
print("The number of duplicate rows after removing is:-",medical_data.duplicated().sum())
```

```python
#3.2 Exercise

#Problem-1
import pandas as pd;
data = pd.read_csv("C:/Users/Natsu Rohit/Downloads/Titanic-Dataset.csv")
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
filtered_data = df[df['Pclass']==1]
print(filtered_data)
```

```python
mean_fare = df['Fare'].mean()
print("The mean fare for the given data is:- ",mean_fare)
median_fare = df['Fare'].median()
print("The median fare for the given data is:- ",median_fare)
max_fare = df['Fare'].max()
print("The maximum fare for the given data is:- ",max_fare)
min_fare = df['Fare'].min()
print("The minimum fare for the given data is:- ",min_fare)
```

```python
#Problem-2
null_count = df['Age'].isnull().sum()
print("The null values contained in the age column of the DataFrame are:- ",null_count)
```

```python
df_without_age = df.dropna(subset=['Age'])
print(df_without_age)
null_count_after = df_without_age['Age'].isnull().sum()
print("The null values contained in the age column of the DataFrame are:- ",null_count_after)
```

```python
#Problem-3
embarked_dummies = pd.get_dummies(df['Embarked'],prefix='Embarked')
df_with_dummies = pd.concat([df, embarked_dummies], axis = 1)
df_with_dummies = df_with_dummies.drop(columns=['Embarked'])
print(df_with_dummies.head())
```

```python
#Problem-4
import matplotlib.pyplot as plt
survival_by_sex = df.groupby('Sex')['Survived'].mean()
print(survival_by_sex)
```

```python
plt.bar(survival_by_sex.index,survival_by_sex.values, color = ['yellow','lightblue'])
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.ylim(0,1)
plt.show()
```

```python
#Problem-5
survival_by_sex_embarked = df_without_age.groupby(['Sex','Embarked'])['Survived'].mean()
print(survival_by_sex_embarked)
```

```python
survival_by_sex_embarked.plot(kind='bar',color=['aqua','maroon'])
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Sex and Port of Embarkation')
plt.ylim(0,1)
plt.show()
```
