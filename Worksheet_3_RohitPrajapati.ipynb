```python
#3.3 Exploratory Data Analysis with Group-by Method Practice Excercise
import pandas as pd
titanic_df = pd.read_csv("C:/Users/Natsu Rohit/Downloads/Titanic-Dataset.csv")
total_fare = titanic_df["Fare"].sum()
print("The total fare amount is:- " ,total_fare)
```

```python
fare_class_1 = titanic_df[titanic_df["Pclass"] == 1]["Fare"].sum()
fare_class_2 = titanic_df[titanic_df["Pclass"] == 2]["Fare"].sum()
fare_class_3 = titanic_df[titanic_df["Pclass"] == 3]["Fare"].sum()
fare_totals = [fare_class_1, fare_class_2, fare_class_3]
fare_proportions = [fare/total_fare for fare in fare_totals]
fare_percentages = [float(round(p*100, 2)) for p in fare_proportions]
print("The fare percentages are:- " ,fare_percentages)
```

```python
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df["Age"].median())
def age_group(age) :
    if age < 10:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"
titanic_df["age_group"] = titanic_df["Age"].apply(age_group)
total_passengers = len(titanic_df)
group_counts = titanic_df["age_group"].value_counts()
group_percentages = (group_counts / total_passengers) * 100
group_percentages.round(2)
```
